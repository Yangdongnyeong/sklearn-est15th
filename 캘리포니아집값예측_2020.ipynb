```python
import os
# 프로젝트의 홈 디렉토리
HOME = os.getcwd()
# 압축파일이 있는 경로
zipPath = os.path.join(HOME, 'california-house-prices.zip')
# 압축파일을 풀 경로
targetPath = os.path.join(HOME, 'data', 'california-house-prices')

print(zipPath)
print(targetPath)
```

```python
from zipfile import ZipFile

# 압축 해제 실행
with ZipFile(zipPath, 'r') as zip_ref:
    zip_ref.extractall(targetPath)
```
